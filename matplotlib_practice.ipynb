```python
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
#set the x-axis to a list of strings for each month

x_axis= ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#set y-axis to a list of floats as the total fare in US dolalrs accumulated for each month.

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot

plt.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()

fig, ax=plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()

fig=plt.figure()

ax=fig.add_subplot()
```

```python
#create the plot with ax.plt

fig=plt.figure()

ax= fig.add_subplot()

ax.plot(x_axis,y_axis)

```

```python
#create the plot with ax.plt()

ax=plt.axes()

ax.plot(x_axis,y_axis)
```

```python
#create the plot

plt.plot(x_axis,y_axis)

plt.show()
```

```python
#create teh plot and add a label for the legend

plt.plot(x_axis,y_axis, label='Boston')

#create labels fro the x and y axes

plt.xlabel("Date")

plt.ylabel("Fare($)")

#set the y limit to between 0 and 45

plt.ylim(0,45)

#create a title

plt.title("Pyber Fare by Month")

#add a legend

plt.legend()
```

```python
#ax plot from below

fig, ax= plt.subplots()

ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')

ax.legend()

ax.set_title("Pyber Fare by Month")

ax.set_xlabel("Date")

ax.set_ylabel("Fare($)")

ax.set_ylim(0,50)

ax.grid()

ax.marker=("D")

```

```python
#create the plot.

plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#create labels for the x and y axes

plt.xlabel("Date")

plt.ylabel("Fare($)")

#Set the ylimit between 0 and 45

plt.ylim(0,45)

#create a title

plt.title("Pyber Fare by Month")

#add a grid

plt.grid()

#add the legend

plt.legend()
```

```python
#Create a bar plot using matlib method

plt.bar(x_axis, y_axis, color="green", label='Boston')

#create the lable for x and y axes

plt.xlabel=("Date")

plt.ylabel=("Fare($)")

#create the title

plt.title("Pyber Fare by Month")

#add the legend of kain

plt.legend()


```

```python
#create a horizontal bar chart

plt.barh(x_axis, y_axis)
```

```python
#previous chart has Jan on botom, not a good read.  Let's invert the y axis

plt.barh(x_axis, y_axis, color="magenta", label="Boston")

plt.gca().invert_yaxis()

plt.legend()

plt.title("Pyber Fare by Month")

plt.ylabel('Fare($)')

plt.xlabel('Date')




```
