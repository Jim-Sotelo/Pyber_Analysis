```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#X axis is Date/Months
x_axis= ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#Y axis is average fare per ride in those months
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(y_axis, x_axis, c='red', label='Chitown')

plt.legend()

plt.title('Pyber Fare by Month')

plt.xlabel('Fare($)')
plt.ylabel('Date')

plt.gca().invert_yaxis()

plt.xlim(0,50)

```

```python
#bubble chart time

#lets make the bubbles size bigger by creating a list for us, check it out created y= then put plt scatter after to use

y_axis_larger=[]

for data in y_axis:
    
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)

    
```

```python
#instead of using the for/in function can use list comprehenson method

plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
#now let's create with the object oriented method of life

fig, ax= plt.subplots()

ax.scatter(x_axis, y_axis)
```

```python
#now let's do a bubble chart with object oriented

fig, ax=plt.subplots()

ax.scatter(y_axis, x_axis, s=y_axis, c='skyblue', marker='o', alpha=.8, edgecolors='black',linewidths=2, label="Beantown")

ax.legend()

ax.set_title('Pyber Fare by Month')

ax.set_ylabel("Date")

ax.set_xlabel("Fare($)")

ax.set_xlim(0,50)

ax.invert_yaxis()

```
