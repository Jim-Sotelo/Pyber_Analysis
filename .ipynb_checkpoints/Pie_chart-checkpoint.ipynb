```python
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(8,8))
x_axis= ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#let's add exploding values to make July stand out

plt.subplots(figsize=(8,8))

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
#How to change some colors for 12 months, yo

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values= (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')


plt.show()
```

```python
#time for object orientation...again

fig, ax=plt.subplots(figsize=(8,8))

ax.pie(y_axis,labels=x_axis, autopct='%.1f%%',explode=(0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0), shadow=True, startangle=90,
      counterclock=False, colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"])

plt.show()
```
