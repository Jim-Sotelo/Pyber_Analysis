```python
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
#set the x-axis to a list of strings for each month

x_axis= ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
fig, ax=plt.subplots()

ax.bar(x_axis, y_axis)
```

```python
#create a horizontal chart via object oriented method

fig, ax=plt.subplots()

ax.barh(x_axis, y_axis)
```

```python
# if we want to flip around the axis we simply use code below, y first then x

fig, ax=plt.subplots()

ax.barh(x_axis, y_axis, color="cyan", label="Chitown")

ax.legend()

ax.set_title("Chitown")

ax.set_title("Pyber Fare by Month")

ax.set_ylabel("Date")

ax.set_xlabel("Fare($)")

ax.invert_yaxis()
```
